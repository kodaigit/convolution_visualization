```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#input and outputs are numpy arrays 
def func(low, high,input,output): 
    result = np.zeros_like(input)
    mask = (input >= low) & (input <= high)
    
    result[mask] = output

    return result 

#create aliases
func_f = func
func_h = func
```

when you convolve a function with another function, the another function is flipped and shifted by t first. And then, for every t, you multiple the two fucntions and intergrate the result of the multiplication from negative and positive infinity. In this notebook, we visualize process by the multiplication of the functions 

```python
shift_const = 3
inputs  = np.linspace(-10,10,100000)
h_outputs = func_h(-3,3,shift_const+ (-1)*inputs,3) #flipped and shifted by a constant 

plt.figure() 
plt.plot(inputs, h_outputs,label = "Shifted")
plt.plot(inputs,func_h(-3,3,inputs,3),label = "Original")
plt.legend()
plt.show()
```
